```
import pandas as arisla
```

```
ds = arisla.read_csv('/content/netflix_titles.csv')
```

```
ds.head()
Filmes_Netflix = ds.dropna(subset=['director'])
Filmes_Netflix = Filmes_Netflix[Filmes_Netflix['type'] == 'Movie']
```

```
oscar = arisla.read_csv('/content/the_oscar_award.csv')
```

```
filtro_director = oscar['category'].str.startswith('DIRECTING')
df_director = oscar[filtro_director]
```

```
BaseFinal = arisla.merge(Filmes_Netflix, df_director, left_on='director', right_on='name', how='left')
display(BaseFinal)
```
